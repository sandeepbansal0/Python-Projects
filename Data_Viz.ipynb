```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("ggplot")
sns.set_theme()

df = sns.load_dataset("tips")  # built-in dataset
df.head()
```

```python
sns.histplot(df["total_bill"], kde=True, bins=30)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
sns.kdeplot(df["tip"], fill=True)
plt.title("KDE Plot  Tip")
plt.show()
```

```python
sns.rugplot(df["total_bill"])
plt.title("Rug Plot")
plt.show()
```

```python
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("Boxplot  Total Bill by Day")
plt.show()
```

```python
sns.barplot(data=df, x="day", y="total_bill", estimator=sum)
plt.title("Sum of Bills by Day")
plt.show()
```

```python
sns.swarmplot(data=df, x="day", y="total_bill")
plt.title("Swarmplot")
plt.show()
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="smoker", style="time")
plt.title("Scatterplot  Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=df, x="size", y="tip", estimator=None)
plt.title("Lineplot  Tip by Party Size")
plt.show()
```

```python
sns.regplot(data=df, x="total_bill", y="tip")
plt.title("Regression Plot")
plt.show()
```

```python
sns.lmplot(data=df, x="total_bill", y="tip", hue="smoker")
```

```python
sns.lmplot(data=df, x="total_bill", y="tip", hue="smoker")
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.jointplot(data=df, x="total_bill", y="tip", kind="hex")
plt.show()
```

```python
g = sns.FacetGrid(df, col="day")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

```python
g = sns.FacetGrid(df, row="sex", col="smoker")
g.map_dataframe(sns.boxplot, x="day", y="total_bill")
```

```python
sns.scatterplot(
    data=df, x="total_bill", y="tip",
    hue="day", size="size", style="sex"
)
plt.show()
```

```python
sns.catplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex",
    kind="box"
)
plt.show()
```

```python
g = sns.JointGrid(data=df, x="total_bill", y="tip")
g.plot(sns.regplot, sns.histplot)
```

```python
grouped = df.groupby("day")["tip"].mean().reset_index()

sns.barplot(data=grouped, x="day", y="tip")
plt.title("Average Tip by Day")
plt.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["total_bill"],
    y=df["tip"],
    mode="markers",
    marker=dict(size=10, color="blue")
))

fig.update_layout(title="Graph Objects Scatter Plot")
fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df["total_bill"], y=df["tip"], mode="markers", name="Tips"))
fig.add_trace(go.Line(x=df["total_bill"], y=df["total_bill"]*0.2, name="20% Rule"))

fig.update_layout(title="Scatter + Line Example")
fig.show()
```

```python
fig.update_layout(
    title="Customized Layout",
    xaxis_title="Total Bill",
    yaxis_title="Tips",
    template="plotly_dark"
)
fig.show()
```

```python
z = np.outer(np.sin(np.linspace(0,2*np.pi,50)), np.cos(np.linspace(0,2*np.pi,50)))

fig = go.Figure(data=[go.Surface(z=z)])
fig.update_layout(title="3D Surface", autosize=True)
fig.show()
```

```python
data = np.random.randn(500)
hist, edges = np.histogram(data, bins=30)

p = figure(title="Histogram", width=700, height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="purple", line_color="white")

show(p)
```
